# Curva de magnetización

```python
import numpy as np
import matplotlib.pyplot as plt

# Nombres de los archivos y etiquetas
archivos = [
    ("magnetizacion32.txt", "N=32"),
    ("magnetizacion64.txt", "N=64"),
    ("magnetizacion128.txt", "N=128"),
]

plt.figure(figsize=(8,6))

for archivo, etiqueta in archivos:
    # Cargar datos: x=temperatura, y=magnetización, dy=error
    datos = np.loadtxt(archivo)
    x = datos[:,0]
    y = datos[:,1]
    dy = datos[:,2]
    plt.errorbar(x, y, yerr=dy, label=etiqueta, marker='o', capsize=3, linestyle='-', markersize=4)

plt.xlabel("T (Temperatura)")
plt.ylabel("M (Magnetización)")
plt.title("Magnetización frente a la Temperatura para distintos tamaños de red\n100000 pasos Monte Carlo")
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo breaksivsno.txt
# Columnas: N, tiempo_con_break, tiempo_sin_break
datos = np.loadtxt("breaksivsno.txt")
N = datos[:, 0]
tiempo_con_break = datos[:, 1]
tiempo_sin_break = datos[:, 2]

# Ajuste lineal para 'con break'
coef_lin_break = np.polyfit(N, tiempo_con_break, 1)
N_fit = np.linspace(N.min(), N.max(), 200)
fit_lin_break = np.polyval(coef_lin_break, N_fit)

# Ajuste cuadrático para 'sin break'
coef_quad_no = np.polyfit(N, tiempo_sin_break, 2)
fit_quad_no = np.polyval(coef_quad_no, N_fit)

plt.figure(figsize=(8,6))
plt.plot(N, tiempo_con_break, 'o', label='Con break')
plt.plot(N_fit, fit_lin_break, '--', label='Ajuste lineal (con break)')
plt.plot(N, tiempo_sin_break, 's', label='Sin break')
plt.plot(N_fit, fit_quad_no, '-', label='Ajuste cuadrático (sin break)')

plt.xlabel("Tamaño de la red (N)")
plt.ylabel("Tiempo de ejecución (segundos)")
plt.title("Comparación del tiempo de ejecución\nAjuste lineal (con break) y cuadrático (sin break)\n $T=1.0$, 100000 pasos Monte Carlo")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Datos de temperatura y magnetización (primeros 7 puntos)
T = np.array([1.5, 2, 2.5])
M = np.array([0.91655, 0.551594, 0.219447])

# Solo usamos M > 0 (en este caso todos lo son)
y_fit = np.log(M)

def modelo(T, c, a, b):
    return c * np.log(a - T) + b

# Valores iniciales para c, a, b
p0 = [1.0, 4.5, 0.0]

popt, pcov = curve_fit(modelo, T, y_fit, p0=p0)
c_opt, a_opt, b_opt = popt
print(f"c = {c_opt:.4f}, a = {a_opt:.4f}, b = {b_opt:.4f}")

plt.scatter(T, y_fit, label="Datos ln[M]")
T_plot = np.linspace(T.min(), a_opt-0.01, 200)
plt.plot(T_plot, modelo(T_plot, *popt), 'r-', label="Ajuste")
plt.xlabel("T")
plt.ylabel("ln[M]")
plt.legend()
plt.show()
```
