```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos
n, PD = np.loadtxt("PD_vs_n.dat", unpack=True)

# Graficar
plt.figure(figsize=(7,4))
plt.plot(n, PD, marker='o', linestyle='-')
plt.xlabel("Paso temporal n")
plt.ylabel("Probabilidad de detección PD(n)")
plt.title("Evolución temporal de la probabilidad de detección")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos
x, V = np.loadtxt("V.dat", delimiter=",", unpack=True)

# Graficar
plt.figure(figsize=(7,4))
plt.plot(x, V, label="V(x)")
plt.xlabel("Posición x")
plt.ylabel("Potencial V(x)")
plt.title("Perfil del potencial")
plt.grid(True)
plt.legend()
plt.show()
```
